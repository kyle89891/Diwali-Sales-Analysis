## **Exploratory Data Analysis**

**Gender**

```python
# plotting a bar chart for Gender and it's count

ax=sns.countplot(x='Gender',data=df)

#for data labels
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

From above two graphs we find out that most of the buyers are female and they have more purchasing power than man.

# **Age**

```python
df.columns
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
#for data labels
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#total amount vs age group
sales_gen=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_gen)
```

From above grpahs we can see that buyers between 26-35 age group(most females) have most purchasing power. 

## **State**

```python
df.columns
```

```python
#total numbers of orders from top 10 states
data2=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by=['Orders'],ascending=False).head(8)
sns.set(rc={'figure.figsize':(15,5)})    #set the widht and heightof chart 
sns.barplot(x='State',y='Orders',data=data2)
```

```python
#total amount from top 10 states
data2=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False).head(8)
sns.set(rc={'figure.figsize':(15,5)})    #set the widht and heightof chart 
sns.barplot(x='State',y='Amount',data=data2)
```

From above charts we know that UP,Maharastra,Karnataka has place more numbers of orders and have more purchasing power.

# **Marital Status**

```python
ax=sns.countplot(data=df,x='Marital_Status')

sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#sales by marital_status
data2=df.groupby(['Marital_Status',"Gender"],as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False).head(8)
sns.set(rc={'figure.figsize':(7,5)})    #set the widht and heightof chart 
sns.barplot(x='Marital_Status',y='Amount',data=data2,hue='Gender')
```

Most of the buyers are married and most of them are womens.


# **Occupation**

```python
ax=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#sales by Occupation
data2=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False)
sns.set(rc={'figure.figsize':(20,5)})    #set the widht and heightof chart 
sns.barplot(x='Occupation',y='Amount',data=data2)
```

Most of the buyers are from IT,Healthcare and Aviation Industry.


# **Product Category**

```python
df.columns
```

```python
ax=sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(30,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#sales by Product_Category
data2=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False).head(10)
sns.set(rc={'figure.figsize':(30,5)})    #set the widht and height of chart 
sns.barplot(x='Product_Category',y='Amount',data=data2)
```

we can say that most of the selling products are from  Food,Clothing and Electronics category.

```python
#top selling products
data2=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by=['Orders'],ascending=False).head(10)
sns.set(rc={'figure.figsize':(30,5)})    #set the widht and height of chart 
sns.barplot(x='Product_ID',y='Orders',data=data2)

```

# **Conclusion**

Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
